```python
import os
import glob
import matplotlib.pyplot as plt

data_path = "/kaggle/input/oral-diseases-images/oral diseases"
diseases = os.listdir(data_path)

class_distribution = {}

# Folder to skip
skip_folder = "Caries_Gingivitus_ToothDiscoloration_Ulcer-yolo_annotated-Dataset"

for disease in diseases:
    # Skip the specified folder
    if disease == skip_folder:
        continue
    
    class_path = os.path.join(data_path, disease)
    num_images = len(glob.glob(os.path.join(class_path, "*.jpg")))
    class_distribution[disease] = num_images

plt.bar(class_distribution.keys(), class_distribution.values())
plt.xlabel('Oral Diseases')
plt.ylabel('Number of Images')
plt.title('Class Distribution of Oral Diseases')

# Rotate x-axis labels vertically
plt.xticks(rotation='vertical')

plt.show()
```

```python
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns

data_path = "/kaggle/input/oral-diseases-images/oral diseases"
diseases = os.listdir(data_path)

# Folder to skip
skip_folder = "Caries_Gingivitus_ToothDiscoloration_Ulcer-yolo_annotated-Dataset"

num_images_per_disease = []

for disease in diseases:
    # Skip the specified folder
    if disease == skip_folder:
        continue
    
    class_path = os.path.join(data_path, disease)
    num_images = len(glob.glob(os.path.join(class_path, "*.jpg")))
    num_images_per_disease.append(num_images)

# Plotting a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=num_images_per_disease)
plt.xlabel('Oral Diseases')
plt.ylabel('Number of Images')
plt.title('Distribution of the Number of Images for Each Oral Disease')
plt.xticks(rotation='vertical')
plt.show()
```
